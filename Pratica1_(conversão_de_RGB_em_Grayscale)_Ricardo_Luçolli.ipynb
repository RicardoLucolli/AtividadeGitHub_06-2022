## Conversão de imagem RGB em imagem Grayscale

(descrição de cada célula de codigo se encontra na célula de texto abaixo da mesma)

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

Este código importa as bibliotecas OpenCV (cv2, usada para processamento de imagens, Numpy (numpy as np, usada para processamento multidimensional de matrizes), matplotlib.pyplot (que faz matplotlib funcionar como MATLAB, sendo matplotlib usado para criação de gráficos), e a última linha possibilita o uso do termo imshow no google collab

```python
#abria a imagem
img = cv2.imread('noot_noot.jpg',1)
cv2_imshow(img)
```

Este código armazena a imagem na variável "img" com a utilização do comando "cv2.imread(path,flag) sendo path o local do arquivo da imagem, e flag a maneira como a imagem será lida, no caso sendo 1 para ler uma imagem colorida.


```python
#mostrando a imagem colorida

cv2_imshow(img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#aplicando conversão básica
#numpy
#img_grayscale_basic = (img[ : , : ,0]+img[ : , : ,1]+img[ : , : ,2])/3

#cv2
B, G, R = cv2.split(img)
img_grayscale_basic = (B+G+R)/3

img_grayscale_basic = np.array(img_grayscale_basic, dtype=np.uint8)
cv2_imshow(img_grayscale_basic)
#cv2.imshow('img_grayscale_basic', img_grayscale_basic)
#cv2.waitKey(0)
#cv2.destroyAllWindows()
```

Neste código a variável img (que contém a imagem) tem seus canais de cor divididos entre as variáveis B, G, e R, que são somados, divididos por 3 e armazenados na variável img_grayscale_basic, que é convertida para o "dtype" uint8, que é o formato que o OpenCV armazena imagens.

```python
#aplicando conversão ponderada
#img_grayscale_basic = 0.299*img[ : , : ,0] + 0.587*img[ : , : ,1] + 0.114*img[ : , : ,2]

#cv2
B, G, R = cv2.split(img)
img_grayscale_pondered = 0.299*B+0.587*G+0.114*R

img_grayscale_pondered = np.array(img_grayscale_pondered, dtype=np.uint8)
cv2_imshow(img_grayscale_pondered)
#cv2.imshow('img_grayscale_basic', img_grayscale_pondered)
#cv2.waitKey(0)
#cv2.destroyAllWindows()
```

Esse código é quase igual ao código da célula acima, com a diferênça sendo no armazenamento dos canais de cor em uma cor apenas, com uso da fórmula para determinar brilho perceptível de cor RGB em vez da somatória e divisão demonstrada na célula anterior, assim a imagem fica com uma qualidade melhor
